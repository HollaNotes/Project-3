```python
# Import dependencies
import pandas as pd
import numpy as np
```

```python
# Read the data into a Pandas DataFrame
realEstateCa_df = pd.read_csv('Resources/RealEstate_California.csv')
realEstateCa_df.head()
```

```python
# Brief summary of the realEstateCa_df DataFrame before cleaning.
realEstateCa_df.info()
```

```python
# Get the crowdfunding_info_df columns.
print(realEstateCa_df.columns)
```

```python
# Get count of unique ids
unique_ids = realEstateCa_df["id"].nunique()
unique_ids
```

```python
realEstateCa_df.dropna(inplace = True)
realEstateCa_df.info()
```

```python
print(realEstateCa_df.duplicated())
```

```python
# Remove homes listed for less than $1000
filter = (realEstateCa_df[realEstateCa_df["price"] >= 1000]) 

# Remove homes with less than 120 living area
filter2 = (filter[filter["livingArea"] >= 120]) 

# Remove homes with no bathrooms
filter3 = (filter2[filter2["bathrooms"] != 0]) 

filter4 = (filter3[filter3["homeType"] != "MULTI_FAMILY"])  

filter5 = (filter4[filter4["homeType"] != "LOT"])  

filter6 = (filter5[filter5["homeType"] != "MILY"])  


```

```python
# Export categories_df and subcategories_df as CSV files.


filter6.to_csv("Resources/filter6.csv", index=False)

filter6.to_json("Resources/JSON_filter6.json") 
```
